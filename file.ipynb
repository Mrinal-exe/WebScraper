```python
def get_states(r):
    link = f"https://www.freeandclear.com/mortgage-lenders/{r}-review"
    res = requests.get(link)
    sup = BeautifulSoup(res.text, 'html.parser')
    d = sup.find('div', class_="mapholdcont").text
    lines = d.splitlines()
    for i in lines:
        if i=="":
            lines.remove(i)
    state = lines[0]
    word = "licensed in "

    if word in state:
        index = state.index(word)
        part1, part2 = state[:index], state[index+12:]
    else:
        part1, part2 = state[:], []
    return part2
```

```python
CompStates = []
for f in ramesh:
    CompStates.append(get_states(f))
```
